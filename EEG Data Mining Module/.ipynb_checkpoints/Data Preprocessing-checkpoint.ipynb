```python
import pandas as pd
import numpy as np
```

```python
eegData = pd.read_csv('eegDataCHIRAG.csv')
print(eegData)
```

```python
eegDataFrame = pd.DataFrame(eegData)
eegDataFrame
```

```python
eegDataFrame = eegDataFrame.fillna(0)
eegDataFrame
```

```python
eegDataFrame.describe()
```

```python
sns.violinplot( y=eegDataFrame["attention"] )
sns.violinplot( y=eegDataFrame["meditation"] , palette="summer")
```

```python
eegEsenseData = eegDataFrame.iloc[:,1:9]
sns.kdeplot(eegEsenseData['delta'], shade=True, bw=0.5, color="olive")
sns.kdeplot(eegEsenseData['theta'], shade=True, bw=10, color="blue")
sns.kdeplot(eegEsenseData['lowAlpha'], shade=True, bw=10, color="red")
sns.kdeplot(eegEsenseData['highAlpha'], shade=True, bw=10, color="red")
sns.kdeplot(eegEsenseData['lowBeta'], shade=True, bw=10, color="red")
sns.kdeplot(eegEsenseData['highBeta'], shade=True, bw=10, color="red")
sns.kdeplot(eegEsenseData['lowGamma'], shade=True, bw=10, color="red")
sns.kdeplot(eegEsenseData['highGamma'], shade=True, bw=10, color="red")
```

```python
a4_dims = (15,10)
fig, ax = plt.subplots(figsize=a4_dims)
sns.boxplot(x="attention", y='meditation',data=eegDataFrame, ax=ax)
```
